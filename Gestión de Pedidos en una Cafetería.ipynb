```python
class Inventario:
    def __init__(self):
        self.ingredientes={} #diccionario

    def agregar_ingrediente(self, ingrediente, cantidad):
        if ingrediente in self.ingredientes:
             self.ingredientes[ingrediente] += cantidad
            
        else:
            self.ingredientes[ingrediente] = cantidad

    def actualizar_stock(self, ingrediente, cantidad):
        if  ingrediente in self.ingredientes:
            self.ingredientes[ingrediente] += cantidad
            print(f'Stock actualizado: {ingrediente} - {cantidad}')
        else:
            print(f'El ingrediente no está en el inventario')
    
    def verificar_stock(self, producto):
        for ingrediente, cantidad_necesaria in  producto.ingredientes.items():
            if ingrediente not in self.ingredientes or self.ingredientes[ingrediente] < cantidad_necesaria:
                return False
        return True
        
    def usar_ingredientes(self, producto):
        for ingrediente, cantidad_necesaria in producto.ingredientes.items():
            self.ingredientes[ingrediente] -= cantidad_necesaria
```

```python
class Producto:
    listaproductos=[]
    def __init__(self, nombre, precio, ingredientes=None):
        self.nombre= nombre
        self.precio= precio
        self.ingredientes = ingredientes if ingredientes else {}

    def añadir_producto(self):
        Producto.listaproductos.append(self)
        print(f'¡Nuevo proucto! - {self.nombre} -')

    def detalles(self):
        print(f'-{self.nombre} -- Precio: ${self.precio}')

    def ingredientes(self):
        return self.ingredientes
    
    @classmethod
    def menu(cls):
        if not cls.listaproductos: 
            print('Menú:')
            print('No hay productos')
        else:
            print("Menú:")
            for producto in cls.listaproductos:
             print(f'-{producto.nombre} --${producto.precio}')

```

```python
class Bebida(Producto):
    def __init__(self, nombre, precio, tamaño, tipo, personalizacion=None, ingredientes= None):
        super().__init__(nombre, precio, ingredientes)
        self.tamaño=tamaño
        self.tipo= tipo
        self.personalizacion= personalizacion or []

    def agregar_personalizacion(self, personalizacion):
        self.personalizacion.append(personalizacion)
        
    

    
    
```

```python
class Pedido:
    def __init__(self, cliente):
        self.cliente = cliente
        self.estado = "pendiente"
        self.total = 0
        self.productos = []

    def agregar_pedido(self, producto, cantidad, inventario):
        if inventario.verificar_stock(producto):
            self.productos.append(producto)
            self.total += producto.precio * cantidad

            inventario.usar_ingredientes(producto)
            print(f'Pedido agregado {cantidad}- {producto.nombre}')
        else:
            print(f'No hay suficiente stock para el producto: {producto.nombre}')


    def calcular_total(self):
        return self.total

    def actualizar_estado(self, estado):
        self.estado = estado
```

```python
class Persona:
    listapersonas= []

    def __init__(self, nombre, correo):
        self.nombre= nombre
        self.correo= correo
        
    def registrar(self):
        Persona.listapersonas.append(self)
        print(f'Se registró correctamente a {self.nombre} con correo de contacto {self.correo}')
    
    def cambiarnombre(self, nombrenuevo):
        self.nombre=nombrenuevo
        print(f'Se ha actualizado el nombre a: {self.nombre}')
    def cambiarcorreo(self, correonuevo):
        self.correo= correonuevo
        print(f'Se ha actualizado el correo a: {self.correo}')

    @classmethod
    def personas_registradas(cls):
        print("Personas registradas:")
        for persona in cls.listapersonas:
            print(f"-{persona.nombre} - {persona.correo}")
```

```python
class Cliente(Persona):
    def __init__ (self, nombre, correo):
        super().__init__(nombre, correo)
        self.historial_pedidos= []

    def realizar_pedido(self, pedido):
        self.historial_pedidos.append(pedido)
        print(f'Se ha realizado un pedido de  {self.pedido.cantidad} {self.pedido.producto}')
       
    def detalles_de_compra(self, pedido):
        print(f'Detalles del pedido de {self.nombre}:')
        for articulo in pedido.productos:
            producto= articulo['producto']
            cantidad= articulo['cantidad']

            print(f'-{cantidad} - {producto.nombre} -- {producto.precio} c/u')
            print(f'Total a pagar: {pedido.calcular_total()}')    
    
    def imprimir_historial(self):
        print(f'Historial de {self.nombre}:')
        if not self.historial_pedidos:
            print("No has hecho ningún pedido")
        else:
            for pedido in self.historial_pedidos:
                print(f'- {pedido.nombre} - {pedido}')
            
```

```python
class Empleado(Persona):
    def __init__(self, nombre, correo, rol):
        super().__init__(nombre, correo)
        self.rol= rol
    
    def actualizar_inventario(self, inventario, ingrediente, cantidad):
        if self.rol== "gerente" or self.rol== "barista":
            inventario.actualizar_stock(ingrediente, cantidad)
            print(f'Se ha actualizado el stock')
            
        else:
            print(f'No tiene permiso para actualizar el inventario')
        
```

```python
class Promocion:
    def __init__(self, descuento):
       
        self.descuento = descuento

    def aplicar_descuento(self, pedido): 
        if pedido.total== 150:
            descuento= self.descuento / 100
            descuento_aplicado= pedido.total * descuento
            pedido.total -= descuento_aplicado
            print(f'Se ha aplicado un descuento de {self.descuento}% ')
            print(f'Total del pedido: {pedido.total:.2f}')
        else:
            print('No se aplicó ningun descuento')
```
